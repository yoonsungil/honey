```python
import os
import librosa
import joblib
import numpy as np
```

```python
from tqdm import tqdm
```

```python
file_path = 'C:/wav/'
file_names = os.listdir(file_path)
files = []
for file_name in file_names:
    files.append(file_path + file_name)
print('저장완료')
```

```python
dataset = []
label=[]
for file in tqdm(files):
    signal, sr = librosa.load(file, sr=16000)
    mfcc = librosa.feature.mfcc(signal, sr, n_fft=400, hop_length=160, n_mfcc=36)
    dataset.append(mfcc.reshape(-1))
    label.append((os.path.basename(file)).split('_')[0])
print('전처리완료')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(label)
y= le.transform(label)
```

```python
le.classes_
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset, y, test_size = 0.2, random_state = 42, shuffle =True)
```

```python
clf = SVC(C=10, kernel = 'rbf', probability=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
joblib.dump(scaler, 'scaler.pkl')
```

```python
len(X_train_scaled[1])
```

```python
clf.fit(X_train_scaled, y_train)
```

```python
print(clf.score(X_test_scaled,y_test))
```

```python
joblib.dump(clf, 'model.pkl')
```

```python
fp = 'data/5차년도_2차/'
wav_id = '5f3ea1919dd513131eace41a.wav'
fname = fp+wav_id
```

```python
from pydub import AudioSegment, silence
```

```python
myaudio = AudioSegment.from_wav(fname)
dBFS = myaudio.dBFS
silence_section = silence.detect_silence(myaudio, min_silence_len = 1000, silence_thresh = dBFS - 16)
silence_section = [((start/1000),(stop/1000)) for start, stop in silence_section]
print(silence_section)
```

```python
section_list = []
for start, end in silence_section:
    section_list.extend([start, end])
section_list = section_list[1:-1]
print(section_list)
```

```python
new_section = []
for i, section in enumerate(section_list):
    if i % 2 == 0:
        new_section.append([section, section_list[i+1]])
```

```python
signal, sr = librosa.load(fname, sr = 16000)
print(signal)
```

```python
reform_signal = []
for start, end in new_section:
    reform_signal.extend(signal[int(start * 16000): int(end * 16000)])
reform_signal = np.array(reform_signal)
```

```python
audio_len = int(librosa.get_duration( y =signal, sr = sr))
print(audio_len)
```

```python
for i in range(0, audio_len -1):
    emotions = []
    sliced_signal = signal[i*16000 : (i+2) *16000]
    mfcc = librosa.feature.mfcc(y = sliced_signal, sr=16000, n_fft=400, hop_length=160, n_mfcc=36)
    mfcc = mfcc.reshape(-1)
    #scaler.fit(X_train)
    mfcc = scaler.transform([mfcc])
    result = clf_from_joblib.predict(mfcc)   #'angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise'
    if result[0] == 0:
        emotion = 'angry'
    elif result[0] == 1:
        emotion = 'disgust'
    elif result[0] == 2:
        emotion = 'fear'
    elif result[0] == 3:
        emotion = 'happy'
    elif result[0] == 4:
        emotion = 'neutral'
    elif result[0] == 5:
        emotion = 'sad'
    elif result[0] == 6:
        emotion = 'surprise'
    print(emotion)
    emotions.append(emotion)
print(max(emotions, key=emotions.count))
```
