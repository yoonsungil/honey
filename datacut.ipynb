```python
import soundfile as sf
import librosa
import numpy as np
import sys
import os
from imutils import paths
from tqdm import tqdm
```

```python
folder = 'data/wav'
files = paths.list_files(folder)

dataset = []
name = []
```

```python
for i in tqdm(files):
    y, sr = librosa.load(i, sr=16000)
    dataset.append([y,sr])
    name.append((os.path.basename(i)))
```

```python
duration = []
for data in dataset:
    duration.append(int(librosa.get_duration(data[0], data[1])))
```

```python
folder1= 'C:/wav/'

for data,nam,duratio in zip(dataset, name, duration):
    y, sr = data
    for i in range(0,duratio-1):
        cutted_y = y[i*16000:(i+2)*16000]
        cutted_name = nam[:-4]
        sf.write(folder1 + cutted_name + 'a' + str(i+1) + '.wav', cutted_y, sr, 'PCM_16')
```
